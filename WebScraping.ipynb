```python
import requests
from bs4 import BeautifulSoup
import random
import sys
import zipfile
sys.setrecursionlimit(1500)
       
def scrapeAsoiaf(url, count):
    response = requests.get(
        url=url,
    )
    soup = BeautifulSoup(response.content, 'html.parser')

    allLinks = soup.find(id="bodyContent").find_all("a")

    random.shuffle(allLinks)
    linkToScrape = 0

    for links in allLinks:
        if links['href'].find("/index.php/") == -1:   
            continue
        if links['href'].find("Template") != -1:
            continue
        if links['href'].find("User:") != -1:
            continue
        if links['href'].find("User_talk") != -1:
            continue
        if links['href'].find("Categor") != -1:
            continue
        if links['href'].find('File:') != -1:
            continue   
        
        linkToScrape = links
        print(links)
        break
        
    newfile = open("index.txt", "a", encoding='utf-8')
    newstring = str(count) + " " + url
    newfile.write(newstring+"\n")
    newfile.close()
    htmlData = soup.find(id="bodyContent").find_all("p")
   
    for htmlLink in htmlData:
        p = htmlLink.get_text() 
        f = open(str(count)+".txt", "a", encoding='utf-8')
        f.write(p)
    f.close()
   
    count +=1
    print(sys.getrecursionlimit())
    if (count > 100):
       
        sys.exit(0)
    scrapeAsoiaf("https://awoiaf.westeros.org" + linkToScrape['href'], count)
    
scrapeAsoiaf("https://awoiaf.westeros.org/index.php/Bran_Stark", 1)  
```

```python
import requests
from bs4 import BeautifulSoup
import random

def scrapeWikiArticle(url):
	response = requests.get(
		url=url,
	)
	
	soup = BeautifulSoup(response.content, 'html.parser')

	title = soup.find(id="firstHeading")
	print(title.text)

	allLinks = soup.find(id="bodyContent").find_all("a")
	random.shuffle(allLinks)
	linkToScrape = 0

	for link in allLinks:
		# We are only interested in other wiki articles
		if link['href'].find("/wiki/") == -1: 
			continue

		# Use this link to scrape
		linkToScrape = link
		break

	scrapeWikiArticle("https://en.wikipedia.org" + linkToScrape['href'])

scrapeWikiArticle("https://en.wikipedia.org/wiki/Web_scraping")
```
